```python
import pandas as pd

# Загрузка данных
file_path = 'bikes_rent.csv'
data = pd.read_csv(file_path)

# Предварительный анализ
print(data.head())
print(data.info())

```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Извлечение необходимых столбцов
X = data[['weathersit']]
y = data['cnt']

# Обучение модели линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Предсказания модели
y_pred = model.predict(X)

# Построение графика
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Weather Situation')
plt.ylabel('Count of Bikes Rented')
plt.title('Linear Regression: Weather Situation vs Count of Bikes Rented')
plt.legend()
plt.show()

print(f"Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
def predict_cnt(weathersit_value):
    df = pd.DataFrame({'weathersit': [weathersit_value]})
    return model.predict(df)[0]

# Пример предсказания
weathersit_value = 2
predicted_cnt = predict_cnt(weathersit_value)
print(f"Predicted count of bikes rented for weather situation {weathersit_value}: {predicted_cnt}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Выбор признаков
X_reduced = data[['temp', 'hum']]
y = data['cnt']

# Обучение модели линейной регрессии
model_reduced = LinearRegression()
model_reduced.fit(X_reduced, y)

# Построение графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_reduced['temp'], X_reduced['hum'], y, color='blue', label='Actual Data')

# Создание сетки для предсказаний
temp_range = np.linspace(X_reduced['temp'].min(), X_reduced['temp'].max(), 100)
hum_range = np.linspace(X_reduced['hum'].min(), X_reduced['hum'].max(), 100)
temp_grid, hum_grid = np.meshgrid(temp_range, hum_range)
cnt_grid = model_reduced.predict(np.c_[temp_grid.ravel(), hum_grid.ravel()]).reshape(temp_grid.shape)

# Поверхность предсказаний
ax.plot_surface(temp_grid, hum_grid, cnt_grid, color='red', alpha=0.5)

ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Count of Bikes Rented')
plt.title('3D Linear Regression: Temperature & Humidity vs Count of Bikes Rented')
plt.legend()
plt.show()
```
