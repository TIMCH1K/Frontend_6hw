```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Загрузка датасета ирисов
iris = load_iris()
X = iris.data

# Определение количества кластеров для метода локтя
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Построение графика метода локтя
plt.figure(figsize=(10, 5))
plt.plot(K, inertia, 'bo-', markersize=8)
plt.xlabel('Количество кластеров, k')
plt.ylabel('Inertia')
plt.title('Метод локтя для определения оптимального количества кластеров')
plt.grid(True)
plt.show()
```

```python
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    silhouette_scores.append(silhouette_score(X, labels))

# Построение графика силуэтного анализа
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores, 'bo-', markersize=8)
plt.xlabel('Количество кластеров, k')
plt.ylabel('Silhouette Score')
plt.title('Силуэтный анализ для определения оптимального количества кластеров')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Загрузка датасета ирисов
iris = load_iris()
X = iris.data

# Функция для вычисления евклидова расстояния
def euclidean_distance(a, b):
    return np.linalg.norm(a - b)

# Функция для инициализации центроидов
def initialize_centroids(X, k):
    np.random.seed(42)
    indices = np.random.choice(len(X), k, replace=False)
    return X[indices]

# Функция для назначения меток кластерам
def assign_clusters(X, centroids):
    clusters = {}
    for i in range(len(X)):
        distances = [euclidean_distance(X[i], centroid) for centroid in centroids]
        cluster = np.argmin(distances)
        if cluster not in clusters:
            clusters[cluster] = []
        clusters[cluster].append(X[i])
    return clusters

# Функция для обновления центроидов
def update_centroids(clusters, k):
    centroids = np.zeros((k, X.shape[1]))
    for cluster in clusters.keys():
        centroids[cluster] = np.mean(clusters[cluster], axis=0)
    return centroids

# Функция для визуализации кластеров
def plot_clusters(X, centroids, clusters, iteration):
    colors = ['r', 'g', 'b', 'y', 'c', 'm']
    plt.figure(figsize=(8, 6))
    for cluster in clusters.keys():
        points = np.array(clusters[cluster])
        plt.scatter(points[:, 0], points[:, 1], s=30, c=colors[cluster], label=f'Cluster {cluster + 1}')
    plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='black', marker='X', label='Centroids')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'Iteration {iteration}')
    plt.legend()
    plt.grid(True)
    plt.show()

# Реализация алгоритма k-means
def kmeans(X, k, max_iterations=100):
    centroids = initialize_centroids(X, k)
    for i in range(max_iterations):
        clusters = assign_clusters(X, centroids)
        plot_clusters(X, centroids, clusters, i + 1)
        old_centroids = centroids
        centroids = update_centroids(clusters, k)
        if np.all(centroids == old_centroids):
            break
    return centroids, clusters

# Вызываем k-means для 3 кластеров
centroids, clusters = kmeans(X, 3)
```
